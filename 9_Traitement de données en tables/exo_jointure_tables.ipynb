Commençons par importer les deux tables dans Python. 

```python
import csv

def import2_csv(nom_fichierCSV): 
    fichier_csv = open(nom_fichierCSV, 'r',encoding = 'UTF-8')
    lecteur_csv= csv.DictReader(fichier_csv, delimiter = ";")
    liste_dicos=[dict(dico_ord) for dico_ord in lecteur_csv]
    fichier_csv.close()
    return liste_dicos

table_films = import2_csv("films.csv")
table_films
```

```python
table_realisateurs = import2_csv("realisateurs.csv")
table_realisateurs
```

### A faire vous-même 3 :

Analyser le code, réinitialiser le noyau puis appliquer la fonction fusion à 

```python
import copy

def fusion(table_de_gauche,table_de_droite):
    table_fusion = []
    for dico_de_gauche in table_de_gauche:
        for dico_de_droite in table_de_droite:
            if dico_de_droite['movie_id'] == dico_de_gauche['movie_id']:
                dico_de_gauche_copie = copy.deepcopy(dico_de_gauche)
                for cle in dico_de_droite:
                    if cle != 'movie_id':
                        dico_de_gauche_copie[cle] = dico_de_droite[cle]
                table_fusion.append(dico_de_gauche_copie)
    return table_fusion
```

```python
fusion(table_films,table_realisateurs)
```
