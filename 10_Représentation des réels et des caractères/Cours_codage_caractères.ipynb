## I. Le codage ASCII

### Prsentation du codage ASCII

Dans les annes 50, il existait un nombre important d'encodages de caractres dans les ordinateurs, les imprimantes ou les lecteurs de carte.<br>
Tous ces encodages taient incompatibles les uns avec les autres, ce qui rendaient les changes particulirement difficiles car il fallait utiliser des programmes pour convertir les caractres d'un encodage dans un autre.
<br>
<br>

<div class='alert-info'>
    
Pour tenter de mettre de l'ordre dans tout a, l'ANSI (*American National Standards Institute*) propose au dbut des annes 60 une norme de codage des caractres appele **ASCII** (pour *American Standard Code for Information Interchange*), permettant de reprsenter l'ensemble des caractres d'un clavier anglophone.

A chaque caractre est associ un nombre binaire sur 8 bits (1 octet).<br> 
En fait, seuls **7 bits** sont utiliss pour coder un caractre, le 8e bit n'est pas utilis pour le codage des caractres.<br> Avec 7 bits il est possible de coder jusqu' $2^7=128$ caractres ce qui est largement suffisant pour un texte crit en langue anglaise (pas d'accents et autres lettres particulires).

## II. Normes ISO 8859
## II.1 Prsentation 
 La norme ASCII convient bien  la langue anglaise, mais pose des problmes dans d'autres langues, par exemple le franais.<br>
 En effet l'ASCII ne prvoit pas d'encoder  certains caractres comme les lettres accentues (, , ,  , , ...),les caractres d'autres langues( , ), les caractres spciaux (, ...) etc... 
 <br>
 <div class='alert-info'>
    
Pour remdier  ces problmes, l'ISO (*Organisation Internationale de Normalisation*) a propos la **norme ISO 8859**.<br>
Cette norme reprend les mmes principes que l'ASCII, mais les nombres binaires associs  chaque caractre sont cods sur **8 bits**, ce qui permet d'encoder jusqu' 256 caractres.

Cependant, malgr un nombre de caractres doubl, cela reste insuffisant pour reprsenter tous les caractres utiliss rien que pour les langues latines.

C'est pourquoi la norme 8859 compte seize tables indpendantes notes 8859-1, 8859-2, ..., 8859-16 dont 10 tables concernant les langues latines.<br>
Chacune de ces tables ont les 128 premiers caractres de la table ASCII et sont compatibles entre-elles (i.e les caractres communs de ces tables ont le mme encodage).


Par exemple, pour les caractres en  franais, la table utilise est la table 8859-1 appele **latin-1** ou la table volue (qui inclut en particulier le symbole ) 8859-9 appele **latin-9**
</div>

# III. La norme Unicode et le format UTF-8
<br>
<div class='alert-info'>
    
Pour viter ces problmes d'encodage, en 1991 une nouvelle norme a vu le jour : **Unicode.**

Unicode a pour ambition de rassembler tous les caractres existant afin qu'une personne utilisant Unicode puisse, sans changer la configuration de son traitement de texte,  la fois lire des textes en franais ou en japonais par exemple.

Unicode est uniquement une table qui regroupe tous les caractres existant au monde. <br>
110 000 caractres sont ainsi recenss avec Unicode, la capacit maximale a t fix  $2^{32}= 4 294 967 295 caractres, c'es--dire le plus grand entier non sign reprsentable avec un mot de 32 bits.<br>
Unicode ne s'occupe pas de la faon dont les caractres sont cods dans la machine. <br>
Unicode accepte alors plusieurs systmes de codage appels UTF (*Universal Transformation Format*).<br>
Les plus connus sont les formats  sont : UTF-8, UTF-16, UTF-32. <br>
Le plus utilis, notamment sur le Web, est le  format **UTF-8.**

Pour encoder les caractres Unicode, UTF-8 utilise un nombre variable d'octets : les caractres "classiques" (les plus couramment utiliss) sont cods sur un octet, alors que des caractres "moins classiques" sont cods sur un nombre d'octets plus important (jusqu' 4 octets). <br>
Un des avantages d'UTF-8 c'est qu'il est totalement compatible avec la norme ASCII et encore mieux avec la norme ISO 8859-1 (latin-1) : Les caractres Unicode cods avec UTF-8 ont exactement le mme code que les 256 mmes caractres en latin-1.
