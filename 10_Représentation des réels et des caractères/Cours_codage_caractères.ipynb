### Introduction :
Nous savons qu'un ordinateur est uniquement capable de traiter des données binaires 0 et 1.<br>
Ainsi, nous avons vu qu'un ordinateur était capable d'encoder n'importe nombre entier naturel en utilisant la base 2.

Donc en théorie, pour écrire du texte, une solution serait d'associer chaque caractère à un unique entier.<br>
En pratique, il faut définir auparavant l'ensemble des caractères à coder. On parle de "*charset*", abréviation de l'anglais "*character set*".<br>
Il faut que tous les ordinateurs utilisent également le même encodage pour que les caractères soient reconnus d'une machine à l'autre.

## I. Le codage ASCII

### Présentation du codage ASCII

Dans les années 50, il existait un nombre important d'encodages de caractères dans les ordinateurs, les imprimantes ou les lecteurs de carte.<br>
Tous ces encodages étaient incompatibles les uns avec les autres, ce qui rendaient les échanges particulièrement difficiles car il fallait utiliser des programmes pour convertir les caractères d'un encodage dans un autre.
<br>
<br>

<div class='alert-info'>
    
Pour tenter de mettre de l'ordre dans tout ça, l'ANSI (*American National Standards Institute*) propose au début des années 60 une norme de codage des caractères appelée **ASCII** (pour *American Standard Code for Information Interchange*), permettant de représenter l'ensemble des caractères d'un clavier anglophone.

A chaque caractère est associé un nombre binaire sur 8 bits (1 octet).<br> 
En fait, seuls **7 bits** sont utilisés pour coder un caractère, le 8e bit n'est pas utilisé pour le codage des caractères.<br> Avec 7 bits il est possible de coder jusqu'à $2^7=128$ caractères ce qui est largement suffisant pour un texte écrit en langue anglaise (pas d'accents et autres lettres particulières).

## II. Normes ISO 8859
## II.1 Présentation 
 La norme ASCII convient bien à la langue anglaise, mais pose des problèmes dans d'autres langues, par exemple le français.<br>
 En effet l'ASCII ne prévoit pas d'encoder  certains caractères comme les lettres accentuées (é, è, à, ñ , Ö, ...),les caractères d'autres langues( β, 漢), les caractères spéciaux (Ø, €...) etc... 
 <br>
 <div class='alert-info'>
    
Pour remédier à ces problèmes, l'ISO (*Organisation Internationale de Normalisation*) a proposé la **norme ISO 8859**.<br>
Cette norme reprend les mêmes principes que l'ASCII, mais les nombres binaires associés à chaque caractère sont codés sur **8 bits**, ce qui permet d'encoder jusqu'à 256 caractères.

Cependant, malgré un nombre de caractères doublé, cela reste insuffisant pour représenter tous les caractères utilisés rien que pour les langues latines.

C'est pourquoi la norme 8859 compte seize tables indépendantes notées 8859-1, 8859-2, ..., 8859-16 dont 10 tables concernant les langues latines.<br>
Chacune de ces tables ont les 128 premiers caractères de la table ASCII et sont compatibles entre-elles (i.e les caractères communs de ces tables ont le même encodage).


Par exemple, pour les caractères en  français, la table utilisée est la table 8859-1 appelée **latin-1** ou la table évoluée (qui inclut en particulier le symbole €) 8859-9 appelée **latin-9**
</div>



## II.2 Les problèmes d'encodage liés à la norme ISO 8859

**Premier problème :** Prenons l'exemple d'un fichier texte en français. Si l'on utilise la norme ISO 8859, il sera encodé en latin-1.<br>
Imaginons que ce même fichier soit ouvert par un grec. S'il utilise la norme ISO 8859, son logiciel de lecture de fichier utilisera d'après le tableau ci-dessus l'encodage 8859-7.<br>
Certains caractères seront sans doute remplacés par d'autres : des problèmes d'encodage risquent de survenir.


**Deuxième problème :** Il existe beaucoup d'autres langues dans le monde qui n'utilisent pas la norme ISO 8859, par exemple le chinois ou le japonais ! <br>
Ainsi, d'autres normes ont donc dû voir le jour, par exemple la norme "GB2312" pour le chinois simplifié ou encore la norme "JIS_X_0208" pour le japonais.

Cette multiplication des normes a très rapidement posé problème. <br>
Imaginons un français qui parle le japonais. Son traitement de texte est configuré pour reconnaitre les caractères de l'alphabet "latin-1" (norme ISO-8859-1).<br>
Un ami japonais lui envoie un fichier texte écrit en japonais. Le français devra modifier la configuration de son traitement afin que ce dernier puisse afficher correctement l'alphabet japonais. S'il n'effectue pas ce changement de configuration, il verra s'afficher des caractères ésotériques comme Ã,  ©

# III. La norme Unicode et le format UTF-8
<br>
<div class='alert-info'>
    
Pour éviter ces problèmes d'encodage, en 1991 une nouvelle norme a vu le jour : **Unicode.**

Unicode a pour ambition de rassembler tous les caractères existant afin qu'une personne utilisant Unicode puisse, sans changer la configuration de son traitement de texte, à la fois lire des textes en français ou en japonais par exemple.

Unicode est uniquement une table qui regroupe tous les caractères existant au monde. <br>
110 000 caractères sont ainsi recensés avec Unicode, la capacité maximale a été fixé à $2^{32}= 4 294 967 295 caractères, c'es-à-dire le plus grand entier non signé représentable avec un mot de 32 bits.<br>
Unicode ne s'occupe pas de la façon dont les caractères sont codés dans la machine. <br>
Unicode accepte alors plusieurs systèmes de codage appelés UTF (*Universal Transformation Format*).<br>
Les plus connus sont les formats  sont : UTF-8, UTF-16, UTF-32. <br>
Le plus utilisé, notamment sur le Web, est le  format **UTF-8.**

Pour encoder les caractères Unicode, UTF-8 utilise un nombre variable d'octets : les caractères "classiques" (les plus couramment utilisés) sont codés sur un octet, alors que des caractères "moins classiques" sont codés sur un nombre d'octets plus important (jusqu'à 4 octets). <br>
Un des avantages d'UTF-8 c'est qu'il est totalement compatible avec la norme ASCII et encore mieux avec la norme ISO 8859-1 (latin-1) : Les caractères Unicode codés avec UTF-8 ont exactement le même code que les 256 mêmes caractères en latin-1.
